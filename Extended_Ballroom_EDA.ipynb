```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import sklearn.metrics as skm
import sklearn.model_selection as skms
import sklearn.preprocessing as skp
import random
import librosa, IPython
import librosa.display as lplt
seed = 12
np.random.seed(seed)
```

```python
df = pd.read_csv('ballroom_extended.csv')
df.head()
```

```python

groups = df.groupby(["label"])['id'].count()

groups = pd.DataFrame({'label':groups.index, 'Number of tracks':groups.values})

groups.head()
groups = groups.sort_values(['Number of tracks'],ascending=False)
groups["label"].replace({0:"Chacha", 1:"Foxtrot", 2:"Jive", 3:"Pasodoble",  4:"Rumba", 5:"Salsa", 6:"Samba", 7:"Slowwaltz",
                        8:"Tango", 9:"Viennesewaltz", 10:"Waltz", 11:"Wcswing", 12:"Quickstep"}, inplace=True)
# groups.plot.bar(color="blue")
sns.set()
plt.figure(figsize = (30, 20))
sns.color_palette("tab10")

ax = sns.barplot(x='label', y='Number of tracks', data=groups, color='blue')
# plt.legend( bbox_to_anchor=(0.95, 0.83), fontsize=40)

plt.xticks(fontsize = 30, rotation=70)
plt.yticks(fontsize = 20)

plt.xlabel('Genre type', fontsize=30)
plt.ylabel('Number of tracks', fontsize=30)
plt.savefig("unbalanced_data_bar_extended_ballroom.png")
plt.show()
```
